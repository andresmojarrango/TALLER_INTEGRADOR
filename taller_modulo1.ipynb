**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
# Promedio
total = 0
for valor in esperanzas_de_vida:
    total += valor
promedio = total / len(esperanzas_de_vida)
print("Promedio:", promedio)

# Máximo
maximo = esperanzas_de_vida[0]
for valor in esperanzas_de_vida:
    if valor > maximo:
        maximo = valor
print("Máximo:", maximo)

# Mínimo
minimo = esperanzas_de_vida[0]
for valor in esperanzas_de_vida:
    if valor < minimo:
        minimo = valor
print("Mínimo:", minimo)

````

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí
import pandas as pd

df = pd.read_csv("salud_mundial.csv")

# Limpieza previa (ejercicio 2.2)
columnas_numericas = df.select_dtypes(include='float64').columns
for col in columnas_numericas:
    df[col] = df[col].fillna(df[col].mean())

# 1. Esperanza de vida promedio por región
ev_region = df.groupby('region')['esperanza_vida'].mean().sort_values(ascending=False)
print(ev_region.round(2))

# 2. Gasto en salud promedio por nivel_ingresos
orden = ['Low', 'Medium', 'High', 'Very High']
gasto_ingresos = df.groupby('nivel_ingresos')['gasto_salud_usd'].mean().reindex(orden)
print(gasto_ingresos.round(2))

# 3. Número de países por región
paises_region = df.groupby('region')['pais'].count().sort_values(ascending=False)
print(paises_region)

# 4. Mayor y menor
print(f"Mayor: {ev_region.idxmax()} — {ev_region.max():.2f} años")
print(f"Menor: {ev_region.idxmin()} — {ev_region.min():.2f} años")
````

### 2.6 — Nueva columna calculada

🔧 **Ejercicio 2.6**
1. Crea una nueva columna llamada `categoria_vida` aplicando la función `clasificar_esperanza_vida` que creaste en la Parte 1 a la columna `esperanza_vida`
2. ¿Cuántos países caen en cada categoría? (usa `value_counts()`)
3. ¿Qué porcentaje del total representa cada categoría?

````python
# Tu código aquí
import pandas as pd

df = pd.read_csv("salud_mundial.csv")

# Limpieza previa (ejercicio 2.2)
columnas_numericas = df.select_dtypes(include='float64').columns
for col in columnas_numericas:
    df[col] = df[col].fillna(df[col].mean())

# Función del ejercicio 1.3
def clasificar_esperanza_vida(años):
    if años >= 75:
        return "Alta"
    elif años >= 60:
        return "Media"
    else:
        return "Baja"

# 1. Crear nueva columna con .apply()
df['categoria_vida'] = df['esperanza_vida'].apply(clasificar_esperanza_vida)
print(df[['pais', 'esperanza_vida', 'categoria_vida']].head(8))

# 2. Conteo por categoría
conteo = df['categoria_vida'].value_counts()
print(conteo)

# 3. Porcentaje
porcentaje = df['categoria_vida'].value_counts(normalize=True).mul(100).round(2)
print(porcentaje)
````

**Escribe tu pregunta:**

¿Los países con mayor urbanización tienen mayor obesidad?



````python
# Tu código aquí
import pandas as pd

df = pd.read_csv("salud_mundial.csv")
columnas_numericas = df.select_dtypes(include='float64').columns
for col in columnas_numericas:
    df[col] = df[col].fillna(df[col].mean())

# Correlación general
corr = df['poblacion_urbana_pct'].corr(df['obesidad_pct'])
print(f"Correlación urbanización vs obesidad: {corr:.4f}")

# Dividir en cuartiles de urbanización
df['cuartil_urbano'] = pd.qcut(df['poblacion_urbana_pct'], q=4,
    labels=['Q1 Poco urbano', 'Q2', 'Q3', 'Q4 Muy urbano'])

resumen = df.groupby('cuartil_urbano', observed=True).agg(
    paises        = ('pais', 'count'),
    urbanizacion  = ('poblacion_urbana_pct', 'mean'),
    obesidad_media= ('obesidad_pct', 'mean')
).round(2)
print(resumen)

# Casos extremos
print(df.nlargest(5, 'poblacion_urbana_pct')[['pais','poblacion_urbana_pct','obesidad_pct']])
print(df.nsmallest(5,'poblacion_urbana_pct')[['pais','poblacion_urbana_pct','obesidad_pct']])
````

### 3.3 — Buenas prácticas

**Responde:**

**¿Qué archivos debería incluir un `.gitignore` en un proyecto de Data Science? Menciona al menos 3 y explica por qué.**

> Tu respuesta:
    1. __pycache__/ y *.pyc
    Python genera estos archivos automáticamente al ejecutar código.
    Son temporales, cambian constantemente y no aportan nada al proyecto —
    solo generan ruido en el historial de commits.

     2. .env
    Contiene variables de entorno como API keys, contraseñas o tokens.
    Subir este archivo a GitHub sería un grave problema de seguridad,
    ya que cualquier persona podría ver tus credenciales.

     3. *.csv / data/
    Los archivos de datos pueden pesar cientos de MB o GB.
    Git no está diseñado para versionar datos pesados — para eso
    existen herramientas como DVC o simplemente se comparten por Drive.

     4. .ipynb_checkpoints/
    Jupyter crea esta carpeta automáticamente como respaldo temporal.
    No es parte del código real y genera conflictos innecesarios
    cuando varias personas trabajan en el mismo notebook.

     5. venv/ / env/
    La carpeta del entorno virtual contiene miles de archivos de
    librerías instaladas. Cada quien debe crear su propio entorno
    local — lo que sí se versiona es requirements.txt.


**¿Cuál es la diferencia entre `git add .` y `git add nombre_archivo`?**

> Tu respuesta:
git add .
    Añade AL STAGING AREA todos los archivos nuevos o modificados
    en la carpeta actual y subcarpetas.
    ES útil cuando quieres hacer commit de todo el trabajo de una vez.
   

 git add nombre_archivo   (ej: git add notebook.ipynb)
    Añade ÚNICAMENTE ese archivo específico al staging area.
    Es útil cuando has modificado varios archivos pero solo quieres
    hacer commit de uno en particular, manteniendo los demás
    fuera del commit por ahora.
